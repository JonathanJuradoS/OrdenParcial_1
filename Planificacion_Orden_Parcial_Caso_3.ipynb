## Objetivo
Aplicar la planificaci贸n de orden parcial para representar las acciones, restricciones de orden y v铆nculos causales.

### Descripci贸n
En una cl铆nica inteligente, la IA debe planificar la atenci贸n de un paciente: registro, toma de signos, evaluaci贸n m茅dica, y entrega de receta.

### Acciones Definidas
- Registrar paciente
- Tomar signos
- Evaluaci贸n m茅dica
- Entregar receta

### Planificaci贸n de orden parcial
Representaremos:
- As: conjunto de acciones
- Os: ordenaciones parciales
- Ls: v铆nculos causales


```python

import networkx as nx
import matplotlib.pyplot as plt

acciones = {
    's1': 'Registrar paciente', 's2': 'Tomar signos', 's3': 'Evaluaci贸n m茅dica', 's4': 'Entregar receta'
}

relaciones = [('s1', 's3'), ('s2', 's3'), ('s3', 's4')]

G = nx.DiGraph()
G.add_nodes_from(acciones.keys())
G.add_edges_from(relaciones)

plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightgreen', node_size=2500, arrows=True)
nx.draw_networkx_labels(G, pos, labels=acciones, font_size=12)
plt.title("Planificaci贸n de Orden Parcial")
plt.show()
```


## Explicaci贸n del c贸digo

1. **Acciones**: Definidas como claves (`s1`, `s2`, etc.) y descripciones.
2. **Relaciones de orden**: Representan los requisitos previos entre acciones.
3. **Grafo dirigido**: Visualiza la planificaci贸n de tareas como un flujo con dependencias.


```python

acciones_iniciales = [n for n in G.nodes if G.in_degree(n) == 0]
print("Acciones iniciales (pueden ejecutarse en paralelo al inicio):")
for a in acciones_iniciales:
    print(f"- {acciones[a]}")

paralelismo_inicial = len(acciones_iniciales)
print(f"Nivel de paralelismo inicial: {paralelismo_inicial}")
```


## Conclusiones

- El modelo permite visualizar los pasos obligatorios en la atenci贸n m茅dica automatizada.
- Acciones como registro y signos pueden realizarse en paralelo.
- La planificaci贸n de orden parcial facilita optimizar flujos en cl铆nicas con IA.
